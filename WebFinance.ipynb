Web Scraping Financial data from Yahoo Finance. 

```python
# Import relevant libraries 
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request as ur
```

```python
# Enter the stock symbol
symbol = 'MSFT'

# URL Links 
url_is = 'https://finance.yahoo.com/quote/' + symbol + '/financials?p=' + symbol
url_bs = 'https://finance.yahoo.com/quote/' + symbol + '/balance-sheet?p=' + symbol
url_cf = 'https://finance.yahoo.com/quote/' + symbol + '/cash-flow?p=' + symbol
```

```python
# Open and Read the link
read_data = ur.urlopen(url_is).read()

# Create the SOUP object
soup_is = BeautifulSoup(read_data, 'lxml')
```

```python
ls = [] # Create empty list

for l in soup_is.find_all('div'):
    ls.append(l.string) 
    # add each element one by one to the list
 
exclude = ('Operating Expenses', 'Net Non Operating Interest Income Expense', 'Other Income Expense', 'Net Income Common Stockholders')# Exclude those columns

ls = [e for e in ls]

# Filter out 'None' elements 
new_ls = list(filter(None,ls))

# After the 12th position
new_ls = new_ls[12:]
new_ls
```

```python
# Iterate 6 items of the list at a time
# it = [iter(new_ls)] * 6
# Create a tuple from the 6 6 items 
is_data = list(zip(*[iter(new_ls)] * 6))
is_data
```
